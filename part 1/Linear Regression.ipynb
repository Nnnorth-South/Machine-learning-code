```python
import numpy as np
from sklearn.linear_model import LinearRegression
import time
```

```python
n_features = 2    #特征数
n_samples  = 100000  #样本数
```

```python
np.random.seed(2020)
w = np.random.normal(0,1,(1,n_features))  #shape: (1,3) , (1,n_features)
b = np.random.normal(0,1,1)               #shape: (1,)

#真实的权重
w_true = np.array([[12,33]])
b_true = np.array([5])
```

```python
w,b
```

```python
np.random.seed(2021)
x = np.random.normal(0,1,(n_samples,n_features)) #x -> shape  (n_samples,n_features)  
y_true = w_true.dot(x.T) + b_true   #y_true -> shape (1,n_samples)
```

```python
#利用初始化权重预测的结果
y_pre = w.dot(x.T) + b
y_sk=y_true.flatten()
#均方根误差
error = 1/2*np.mean((y_pre-y_true)**2)
```

```python
y_true,y_sk
```

```python
##error,w,b
```

```python
start=time.time()
m = 10          #用于更新的样本数
epochs = 1000   #迭代的总次数
lr = 0.1       #学习率
err_list = []
e = 1
w_list=[]
b_list=[]
import random
while 1:
    idx = random.sample(list(range(n_samples)),m) #[xx,xx,xx,...]
    x_temp = x[idx,:]
    w = w - lr*1/m*(-1)*(y_true[:,idx]-y_pre[:,idx]).dot(x[idx,:])
    b = b - lr*1/m*(-1)*(y_true[:,idx]-y_pre[:,idx]).sum()
    w_list.append(w)
    b_list.append(b)
    y_pre = w.dot(x.T) + b
    
    #计算误差
    mse = 1/2*np.mean((y_true-y_pre)**2)
    err_list.append(mse)
    
    e += 1
    if e>epochs:
        end=time.time()
        break
```
